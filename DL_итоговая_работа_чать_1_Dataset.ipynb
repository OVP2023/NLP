```python
from datasets import load_dataset, Dataset
import numpy as np
```

```python
dataset = load_dataset('data-silence/lenta.ru_2-extended') # Загрузка датасета новостей
dataset
```

```python
raw_train_dataset = dataset["train"]
raw_train_dataset[0]
```

```python
# Назначаем нежелательные категории
categories_topic = ["Силовые структуры", "Бывший СССР"]
categories_tags = ["Происшествия", "Политика","Следствие и суд","Криминал"]

# Создаем список заголовков и меток
label=[]
texts = dataset["train"]["title"]
labels_topic = dataset["train"]["topic"]
labels_tags = dataset["train"]["tags"]
for i in range(len(texts)):
    if (labels_topic[i] in categories_topic  or labels_tags[i] in categories_tags):
        label.append(1)
    else:
        label.append(0)
```

```python
import pandas as pd

d = {'Заголовки':texts,'label':label}
df = pd.DataFrame(data=d)
```
